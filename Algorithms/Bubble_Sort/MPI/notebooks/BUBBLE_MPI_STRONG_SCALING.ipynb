# Strong Scaling
2^16

```python
tk16 = th.Thicket.from_caliperreader(glob("CALI_FILES_MPI/*-a65536*.cali"))
```

```python
tk16.dataframe
```

```python
tk16.metadata
```

```python
tk16.statsframe.dataframe["time"] = 1
print(tk16.tree())
```

```python
gb16 = tk16.groupby("SortType")
```

```python
ctk16 = th.Thicket.concat_thickets(
    thickets=list(gb16.values()),
    headers=list(gb16.keys()),
    axis='columns',
    metadata_key='num_procs'
)
```

```python
ctk16.dataframe
```

```python
ctk16.dataframe = ctk16.dataframe.reset_index().drop([("node", ""), ('1% perturbed', "nid"), ('1% perturbed', "spot.channel"),('1% perturbed', "Min time/rank"), ('1% perturbed', "Max time/rank"),('1% perturbed', "Total time"), ('1% perturbed', "Variance time/rank"),
                                                      ('random', "nid"), ('random', "spot.channel"),('random', "Min time/rank"), ('random', "Max time/rank"),('random', "Total time"), ('random', "Variance time/rank"),
                                                      ('reverse', "nid"), ('reverse', "spot.channel"),('reverse', "Min time/rank"), ('reverse', "Max time/rank"),('reverse', "Total time"), ('reverse', "Variance time/rank"),
                                                      ('sorted', "nid"), ('sorted', "spot.channel"),('sorted', "Min time/rank"), ('sorted', "Max time/rank"),('sorted', "Total time"), ('sorted', "Variance time/rank"),
                                                     ], axis=1)
```

```python
ctk16.dataframe = ctk16.dataframe.rename({("name", ""): "name", ("num_procs", ""): "num_procs"}).set_index(["name", "num_procs"])
ctk16.dataframe
```

```python
main_function = ctk16.dataframe.loc["main_function"]
comm = ctk16.dataframe.loc["comm"]
comm_large = ctk16.dataframe.loc["comm_large"]
MPI_gather = ctk16.dataframe.loc["MPI_gather"]
MPI_scatter = ctk16.dataframe.loc["MPI_scatter"]
MPI_sendrecv_replace = ctk16.dataframe.loc["MPI_sendrecv_replace"]
comp_large = ctk16.dataframe.loc["comp_large"]
correct_check = ctk16.dataframe.loc["correct_check"]
data_initialization = ctk16.dataframe.loc["data_initialization"]
```

```python
regions = [main_function, comm, comm_large, MPI_gather, MPI_scatter, MPI_sendrecv_replace, comp_large, correct_check, data_initialization]
names = ["main_function", "comm", "comm_large", "MPI_gather", "MPI_scatter", "comp_large", "correct_check", data_initialization]
```

```python
for region, name in zip(regions, names):
    plt.figure(figsize=(10, 6))  # Adjust the figure size if needed
    legend_labels = []
    for column in region.columns:
        first_index = column[0]  # Extract the first index
        legend_labels.append(first_index)
        plt.plot(region.index, region.xs(column, axis=1), marker='o', label=column)

    plt.xlabel('Number of Processors')
    plt.ylabel('Avg time per rank (s)')
    plt.title(f'Bubble Sort MPI - Strong Scaling for 2^16 Elements: {name}')
    plt.legend(legend_labels)
    plt.grid(True)
    plt.show()
```
