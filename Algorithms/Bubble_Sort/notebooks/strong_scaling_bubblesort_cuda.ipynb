# Strong Scaling
2^16


```python
tk16 = th.Thicket.from_caliperreader(glob("cali_data_cuda/*65536*.cali"))
tk16.dataframe
```

```python
tk16.metadata
```

```python
tk16.statsframe.dataframe["time"] = 1
print(tk16.tree())
```

```python
gb16 = tk16.groupby("SortType")
```

```python
ctk16 = th.Thicket.concat_thickets(
    thickets=list(gb16.values()),
    headers=list(gb16.keys()),
    axis='columns',
    metadata_key='num_threads'
)
```

```python
ctk16.dataframe
```

```python
ctk16.dataframe = ctk16.dataframe.reset_index().drop([("node", ""),
                                                      ('random', "nid"), ('random', "spot.channel"),('random', "Min time/rank"), ('random', "Max time/rank"),('random', "Total time"),                                            
                                                      ('random', 'Avg GPU time/rank'), ('random', 'Min GPU time/rank'), ('random', 'Max GPU time/rank'), ('random', 'Total GPU time'),
                                                      ('reverse', "nid"), ('reverse', "spot.channel"),('reverse', "Min time/rank"), ('reverse', "Max time/rank"),('reverse', "Total time"),                                            
                                                      ('reverse', 'Avg GPU time/rank'), ('reverse', 'Min GPU time/rank'), ('reverse', 'Max GPU time/rank'), ('reverse', 'Total GPU time'),
                                                      ('sorted', "nid"), ('sorted', "spot.channel"),('sorted', "Min time/rank"), ('sorted', "Max time/rank"),('sorted', "Total time"),                                            
                                                      ('sorted', 'Avg GPU time/rank'), ('sorted', 'Min GPU time/rank'), ('sorted', 'Max GPU time/rank'), ('sorted', 'Total GPU time'),
                                                      ('1% perturbation', "nid"), ('1% perturbation', "spot.channel"),('1% perturbation', "Min time/rank"), ('1% perturbation', "Max time/rank"),('1% perturbation', "Total time"),                                            
                                                      ('1% perturbation', 'Avg GPU time/rank'), ('1% perturbation', 'Min GPU time/rank'), ('1% perturbation', 'Max GPU time/rank'), ('1% perturbation', 'Total GPU time')                                                      
                                                     ], axis=1)
```

```python
ctk16.dataframe
```

```python
ctk16.dataframe = ctk16.dataframe.rename({("name", ""): "name", ("num_threads", ""): "num_threads"}).set_index(["name", "num_threads"])
ctk16.dataframe
```

```python
main_function = ctk16.dataframe.loc["main_function"]
comm = ctk16.dataframe.loc["comm"]
comm_large = ctk16.dataframe.loc["comm_large"]
cudaMemoryCopy = ctk16.dataframe.loc["cudaMemoryCopy"]
comp = ctk16.dataframe.loc["comp"]
comp_large = ctk16.dataframe.loc["comp_large"]
correct_check = ctk16.dataframe.loc["correct_check"]
data_initialization = ctk16.dataframe.loc["data_initialization"]
```

```python
regions = [main_function, comm, comm_large, cudaMemoryCopy, comp, comp_large, correct_check, data_initialization]
names = ["main_function", "comm", "comm_large","cudaMemoryCopy", "comp", "comp_large", "correct_check", "data_init"]
```

```python
for region, name in zip(regions, names):
    plt.figure(figsize=(10, 6))  # Adjust the figure size if needed
    legend_labels = []
    for column in region.columns:
        first_index = column[0]  # Extract the first index
        legend_labels.append(first_index)
        plt.plot(region.index, region.xs(column, axis=1), marker='o', label=column)

    plt.xlabel('Number of Threads')
    plt.ylabel('Avg time per rank (s)')
    plt.title(f'Bubble Sort CUDA - Strong Scaling for 2^16 Elements: {name}')
    plt.legend(legend_labels)
    plt.grid(True)
    plt.show()
```
